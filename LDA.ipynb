```python
import warnings
warnings.filterwarnings('ignore')

import os
import gensim
import pyLDAvis
import pandas as pd
from gensim import corpora
from pyLDAvis import gensim as gensimvis
from timeit import default_timer as timer
```

```python
file = 'LSTM.xlsx'
df = pd.read_excel(file, 'Sheet1', header=0, engine='openpyxl')
df
```

```python
texts = df.content.str.split()

dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]

num_tokens = len(dictionary)
print('Number of unique tokens: %d' % num_tokens)

num_doc = len(corpus)
print('Number of documents: %d' % num_doc)
```

```python
import warnings
warnings.filterwarnings('ignore')

NUM_TOPICS = 5
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=NUM_TOPICS, id2word=dictionary, passes=120)
```

```python
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(ldamodel, corpus, dictionary)
pyLDAvis.save_html(vis, 'LDA_topic.html')
pyLDAvis.display(vis)
```
